# Simple cost   :

## $$f(x)=x^2+2*x+3$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2+2*x+3
```

```python
x_1 = np.linspace(start=-10,stop=10,num=50)
plt.xlim([-7,10])
plt.ylim(0,70)
plt.xlabel('x')
plt.ylabel('f(x)')

plt.plot(x_1,f(x_1),c='b',lw=4)
plt.show()
```

```python
def df(x):
    return 2*x+2
```

## slope and derivative

```python
# chart 1
plt.figure(figsize=[15,5])

plt.subplot(1,2,1)
x_1 = np.linspace(start=-10,stop=10,num=50)
plt.xlim([-7,10])
plt.ylim(0,70)
plt.xlabel('x')
plt.ylabel('f(x)')

plt.plot(x_1,f(x_1),c='b',lw=4)


#chart 2
plt.subplot(1,2,2)

plt.grid()
plt.xlim([-10,10])
plt.ylim(-15,20)
plt.xlabel('x',fontsize=15)
plt.ylabel('df(x)',fontsize=15)
plt.title('slope and derivative',fontsize=16)
plt.plot(x_1,df(x_1),c='skyblue')
plt.show()
```
